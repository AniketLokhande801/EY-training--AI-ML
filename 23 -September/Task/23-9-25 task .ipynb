```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
totalrevenueperproduct=df.groupby("Product")["Revenue"].sum()
totalrevenueperproduct
```

```python
avgunitsperproduct=df.groupby("Product")["Units_Sold"].mean()
avgunitsperproduct
```

```python
import seaborn as sns

sns.lineplot(data=df,x="Month",y="Revenue",hue="Product")
plt.title('Monthly Revenue Trend (Janâ€“Mar)')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
totalrevenueperproduct.plot(kind="bar",color="lightblue",edgecolor="black")
plt.title("total of revenue by product")
plt.ylabel("revenue")
plt.xlabel("product")
plt.xticks(rotation=45)
plt.plot()
```

```python
avgunitsperproduct.plot(kind="bar",color="lightblue",edgecolor="black")
plt.title("avg units sold by product")
plt.ylabel("units sold")
plt.xlabel("product")
plt.xticks(rotation=45)
plt.plot()
```
